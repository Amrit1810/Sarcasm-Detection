```python
# Extract stop words
stop = set(stopwords.words("english"))

# Remove punctuation marks
exclude = set(string.punctuation)
```

```python
stop
```

```python
exclude
```

```python
# Text pre processing
def clean(doc):
    stop_free = ' '.join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join([i for i in stop_free if i not in exclude])
    num_free = ''.join(i for i in punc_free if not i.isdigit())
    return (num_free)

post_corpus = [clean(post.iloc[i,2]) for i in range(0, post.shape[0])]
```

```python
# Create document term matrix
tdm = textmining.TermDocumentMatrix()
for i in post_corpus:
        print(i)
        tdm.add_doc(i)
```

## Sentiment Analysis

# Text Blob

```python
# Sentiment analysis using Text Blob
# Create empty dataframe to store results
FinalResults = pd.DataFrame()

# Run Engine
for i in range(0, post.shape[0]):
    
    blob = TextBlob(post.iloc[i,2])
    
    temp = pd.DataFrame({'Comments': post.iloc[i,2], 'Polarity': blob.sentiment.polarity}, index = [1])
    
    FinalResults = FinalResults.append(temp)    
```

```python
FinalResults
```

# Vader

```python
# Sentiment Analysis using Vader
FinalResults_Vader = pd.DataFrame()

# Create engine
analyzer = SentimentIntensityAnalyzer()

# Run Engine
for i in range(0, post.shape[0]):
    
    snt = analyzer.polarity_scores(post.iloc[i,2])
    
    temp = pd.DataFrame({'Comments': post.iloc[i,2], 'Polarity': list(snt.items())[3][1]}, index = [0])

    FinalResults_Vader = FinalResults_Vader.append(temp)
```

# Vader - for Single Sentence

```python
# Sentiment Analysis using Vader
text = input("Text: ")

FinalResults_Vader = pd.DataFrame()
# Create engine
analyzer = SentimentIntensityAnalyzer()
# Run Engine
snt = analyzer.polarity_scores(text)
temp = pd.DataFrame({'Comments': text, 'Polarity': list(snt.items())[3][1]}, index = [0])
FinalResults_Vader = FinalResults_Vader.append(temp)

if FinalResults_Vader['Polarity'][0] < 0:
    print("Sarcasm is not present")
else:
    print("Sarcasm is present")
FinalResults_Vader['Polarity']
```
